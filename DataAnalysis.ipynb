```python
# import das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importar o documento csv
dados = pd.read_csv('/content/Dados.csv')
avaliacoes = pd.read_csv('/content/Avaliacoes.csv')
```

```python
# exibir o cabeçalho da planilha
dados.head()
```

```python
# mostrar as infos da planilha
dados.info()
```

```python
# exibir descrição da coluna
dados.describe()
```

```python
# supondo que o simbolo da rupia seja '₹'
dados['actual_price'] = dados['actual_price'].str.replace('₹', '')
```

```python
# convertendo '.' para espaço ''
dados['actual_price'] = dados['actual_price'].str.replace('.', '')
```

```python
# convertendo ',' para espaço '.'
dados['actual_price'] = dados['actual_price'].str.replace(',', '.')
```

```python
# convertendo a coluna para o tipo numérico(opcional)
dados['actual_price'] = pd.to_numeric(dados['actual_price'])
```

```python
# exibindo os primeiros registros após a remoção do símbolo da moeda
dados.head()
```

```python
# supondo que o simbolo da rupia seja '₹', convertendo '.' para espaço '' e convertendo ',' para espaço '.'
dados['product_price'] = dados['product_price'].str.replace('₹', '').str.replace('.', '').str.replace(',', '.')

# convertendo a coluna para o tipo numérico(opcional)
dados['product_price'] = pd.to_numeric(dados['product_price'])

# exibindo os primeiros registros após a remoção do símbolo da moeda
dados.head()
```

```python
# convertendo ',' por '.' da coluna rating
dados['rating'] = dados['rating'].str.replace(',', '.')

# convertendo a coluna para o tipo numérico(opcional)
dados['rating'] = pd.to_numeric(dados['rating'])

# exibindo os primeiros registros após a conversão
dados.head()
```

```python
# exibir a descrição das colunas no valor numérico
dados.describe()
```

## **Gráfico**

```python
# gráfico de barras para contagem de avalialções acima de 4.5
ratings_above_4_5 = dados [dados['rating'] > 4.5]
rating_counts = ratings_above_4_5['rating'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.title('Contagem de Avaliações acima de 4.5')
plt.xlabel('Avaliação')
plt.ylabel('Contagem')
plt.show()
```

```python
faixa1 = dados[dados['actual_price'] <= 50]
faixa2 = dados[(dados['actual_price'] > 50) & (dados['actual_price'] <= 500)]
faixa3 = dados[dados['actual_price'] > 500]

# contagem de produtos em cada faixa de valor
contagem_faixa1 = len(faixa1)
contagem_faixa2 = len(faixa2)
contagem_faixa3 = len(faixa3)

# preparando os dados para o gráfico de pizza
contagens = [contagem_faixa1, contagem_faixa2, contagem_faixa3]
faixas = ['ate 50', 'entre 50 e 500', 'acima de 500']
explode = (0.1, 0, 0)

# criando o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(contagens, labels=faixas, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Distribuição de Produtos por Faixa de Valor')
plt.axis('equal') # qualiza o aspecto para que o gráfico fique redondo
plt.show()
```

```python
import matplotlib.pyplot as plt

faixa1 = dados[dados['actual_price'] <= 50]
faixa2 = dados[(dados['actual_price'] > 50) & (dados['actual_price'] <= 500)]
faixa3 = dados[dados['actual_price'] > 500]

# contagem de produtos em cada faixa de valor
contagem_faixa1 = len(faixa1)
contagem_faixa2 = len(faixa2)
contagem_faixa3 = len(faixa3)

# preparando os dados para o gráfico de pizza
contagens = [contagem_faixa1, contagem_faixa2, contagem_faixa3]
faixas = ['até 50', 'entre 50 e 500', 'acima de 500']

# Encontra o índice da fatia com o maior valor
indice_maior_valor = contagens.index(max(contagens))

# Cria a lista 'explode' com 1 no índice correspondente e 0 em outros índices
explode = [0.1 if i == indice_maior_valor else 0 for i in range(len(contagens))]

plt.figure(figsize=(8, 8))
plt.pie(contagens, labels=faixas, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Distribuição de Produtos por Faixa de Valor')
plt.axis('equal') # qualiza o aspecto para que o gráfico fique redondo
plt.show()
```
